Example:

     Loop through the set, and print the values:

c) **Change Items in a Set:** 
Once a set is created, you cannot change its items, but you can add new items.

Add Items: To add one item to a set use the add() method. To add more than one item to a set use the update() method.

<br>Example: Add an item to a set, using the add() method: 
<br>thisset = {"apple", "banana", "cherry"} 
<br>thisset.add("orange") 
<br>print(thisset)

   OUTPUT:
   {'apple', 'orange', 'banana', 'cherry'}

Example: Add multiple items to a set, using the update() method: 
<br>thisset = {"apple", "banana", "cherry"} 
<br>thisset.update(["orange", "mango", "grapes"]) 
<br>print(thisset)

To determine how many items a set has, use the len() method.
